# 10. Проверить несколько гипотез для данных, чтобы понять почему сотрудники покидают компанию

сравним последнюю оценку уволенных со средней последней оценкой в генеральной совокупности

используем одновыборочный t-тест Стьюдента

```python
stats.ttest_1samp(left_co['last_evaluation'], hr_data['last_evaluation'].mean())
```

pvalue > 0.05, это значит, что последняя оценка уволившихся не отличается от всей совокупности, что не повлияла на решение сотрудника уволиться

сравним уровень удовлетворенности уволенных со средним уровнем удовлетворенности в генеральной совокупности

используем одновыборочный t-тест Стьюдента

```python
stats.ttest_1samp(left_co['satisfaction_level'], hr_data['satisfaction_level'].mean())
```

pvalue < 0.05, это значит, что уровень удовлетворенности уволившихся отличается от всей совокупности и это означает, что работникам не комфортно работать в компании по каким-то причинам

сравним средние часы в месяц уволенных со средними часами в месяц в генеральной совокупности

используем одновыборочный t тест Стьюдента

```python
stats.ttest_1samp(left_co['average_montly_hours'], hr_data['average_montly_hours'].mean())
```

pvalue < 0.05, это значит, что средние часы в месяц уволившихся отличаются от всей совокупности, может это означает о ненормированном графике

сравним стаж уволенных со средним стажем в генеральной совокупности

используем одновыборочный t тест Стьюдента

```python
stats.ttest_1samp(left_co['time_spend_company'], hr_data['time_spend_company'].mean())
```

pvalue < 0.05, это значит, что стаж уволившихся отличается от среднего стажа всей совокупности
возможно сотрудник, который как правило не задерживается в компании

сравним уровень удовлетворенности уволенных с уровнем удовлетворенности работающих сотрудников

используем двухвыборочный t критерий

```python
stats.ttest_ind(left_co['satisfaction_level'], working['satisfaction_level'])
```

pvalue < 0.05, это значит, что уровень удовлетворенности уволенных отличается от уровня удовлетворенности работающих

сравним стаж в компании уволенных со стажем работающих сотрудников 

используем двухвыборочный t критерий

```python
stats.ttest_ind(left_co['time_spend_company'], working['time_spend_company'])
```

pvalue < 0.05, это значит, что стаж уволенных отличается от стажа работающих

Строим Random Forest

```python
output_y = hr_data.left
input_x = hr_data[['satisfaction_level', 'last_evaluation', 'number_project', 'average_montly_hours', 'time_spend_company', 'Work_accident', 'promotion_last_5years']]
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
targets=(output_y)
print(targets.value_counts())
Y=le.fit_transform(targets)
```

```python
from sklearn.model_selection import train_test_split
X_train_22, X_test_22, y_train_22, y_test_22 = train_test_split(input_x, Y, test_size=0.3)
```

```python
from sklearn.ensemble import RandomForestClassifier

# создаем модель деревья решений
# выбираем 200 деревьев в качестве параметра
model=RandomForestClassifier(n_estimators=200)

# обучаем модель
model.fit(X_train_22,y_train_22)
```

```python
headers = list(X_train_22.columns.values)

feature_imp = pd.Series(model.feature_importances_,index=headers).sort_values(ascending=False)

f, ax = plt.subplots(figsize=(6, 6))
sns.barplot(x=feature_imp, y=feature_imp.index)

plt.xlabel('Градация Значений')
plt.ylabel('Значения')
plt.title("Главные определения")
plt.legend()
plt.show()
```

```python
from sklearn.metrics import classification_report
model_pred = model.predict(X_test_22)

print(classification_report(y_test_22, model_pred))

model.score(X_test_22, y_test_22)
```
