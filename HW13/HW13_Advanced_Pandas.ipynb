```python
data = pd.DataFrame({
    'keyword': ['курс гривны к рублю', 'доллары в рубли', '100 долларов в рублях', 'курс рубля'],
    'shows': [125076, 114173, 97534, 53546],
})
```

```python
import pandas as pd
```

```python
data = pd.DataFrame({
    'keyword': ['курс гривны к рублю', 'доллары в рубли', '100 долларов в рублях', 'курс рубля'],
    'shows': [125076, 114173, 97534, 53546],
})
```

```python
data
```

```python
from pymystem3 import Mystem
```

```python
def find_lemmas(row):
    m = Mystem()
    return ' '.join(m.lemmatize(row ['keyword']))
```

```python
data['lemmas'] = data.apply(find_lemmas, axis=1)
data
```

```python
%%time
data['lemmas'] = data.apply(find_lemmas, axis=1)
data
```
