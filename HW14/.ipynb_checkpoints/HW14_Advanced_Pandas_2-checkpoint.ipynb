Задание 1

Используйте файл с оценками фильмов ml-latest-small/ratings.csv. Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. Под временем жизни понимается разница между максимальным и минимальным значением столбца timestamp для данного значения userId.

```python
import pandas as pd
```

```python
rating = pd.read_csv('ml-latest-small/ratings.csv')
```

```python
rating
```

```python
x = rating.groupby('userId').agg({'timestamp': ['max','min'], 'rating': 'count'})
x
```

```python
x.head()
```

```python
y = x.loc[x.rating['count'] > 100]
y
```

```python
y['user_lifetime'] = y.apply(lambda row: row.timestamp['max'] - row.timestamp['min'], axis=1)
y
```

```python
y.head()
```

```python
y.user_lifetime.mean()
```

```python
y.rating.mean()
```

```python
# ВОПРОС: Константин, не могу понять, а вернее запутался на пустяках, а среднее в MAX и MIN как найти?
# по аналогу предыдушего не получается? (ради интереса) спасибо 

y.timestamp.min.mean()
```

Задание 2

Дана статистика услуг перевозок клиентов компании по типам (см. файл с кодом занятия). 

Необходимо сформировать две таблицы:
- таблицу с тремя типами выручки для каждого client_id без указания адреса клиента
- аналогичную таблицу по типам выручки с указанием адреса клиента

Обратите внимание, что в процессе объединения таблиц данные не должны теряться.

```python
rzd = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115],
        'rzd_revenue': [1093, 2810, 10283, 5774, 981]
    }
)
rzd
```

```python
auto = pd.DataFrame(
    {
        'client_id': [113, 114, 115, 116, 117],
        'auto_revenue': [57483, 83, 912, 4834, 98]
    }
)
auto
```

```python
air = pd.DataFrame(
    {
        'client_id': [115, 116, 117, 118],
        'air_revenue': [81, 4, 13, 173]
    }
)
air
```

```python
client_base = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115, 116, 117, 118],
        'address': ['Комсомольская 4', 'Энтузиастов 8а', 'Левобережная 1а', 'Мира 14', 'ЗЖБИиДК 1', 
                    'Строителей 18', 'Панфиловская 33', 'Мастеркова 4']
    }
)
client_base
```

```python
joined_1 = rzd.merge(auto, how='outer').merge(air, how='outer')
joined_1
```

```python
# len(rzd)
# len(auto)
len(air)
```

```python
joined_2 = client_base.merge(rzd, how='outer').merge(auto, how='outer').merge(air, how='outer')
joined_2
```

```python
# Констанин, а как правильно будет записать, с outer (где данные не теряются) или как ниже, у меня ничено не поменялось) забавно
joined_2 = client_base.merge(rzd, how='left').merge(auto, how='left').merge(air, how='left')
joined_2
```

```python
# ДЛЯ СЕБЯ (ИНТЕРЕСНО)

joined_2 = client_base.merge(rzd, how='left').merge(auto, how='left').merge(air, how='right')
joined_2
```

```python
# ВОПРОС: Константин, при right что-то совсем странное получилось... :-)

joined_2 = client_base.merge(rzd, how='right').merge(auto, how='right').merge(air, how='right')
joined_2
```
